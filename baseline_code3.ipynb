### Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨

````python
import os
from torchvision import transforms, datasets
from torch.utils.data import DataLoader, Subset
from sklearn.model_selection import train_test_split

# ------------------------------------------------------------------------
# 1) ÏõêÎ≥∏ Î∞è Ï¶ùÍ∞ï Transform Ï†ïÏùò
# ------------------------------------------------------------------------
original_transform = transforms.Compose([
    # ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÎ•º (224√ó224)Î°ú Î¶¨ÏÇ¨Ïù¥Ï¶à
    transforms.Resize((224, 224)),
    # ÌÖêÏÑú Î≥ÄÌôò & Ï†ïÍ∑úÌôî
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

augmented_transform = transforms.Compose([
    # ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÎ•º (224√ó224)Î°ú Î¶¨ÏÇ¨Ïù¥Ï¶à
    transforms.Resize((224, 224)),
    # 5Í∞ÄÏßÄ Ï§ë ÌïòÎÇòÎßå ÎûúÎç§ ÏÑ†ÌÉù (ÏïÑÎ¨¥Í≤ÉÎèÑ Ïïà ÌïòÍ∏∞ + 4Í∞ÄÏßÄ Ï¶ùÍ∞ï)
    transforms.RandomChoice([
        transforms.Lambda(lambda x: x),  # (1) ÏïÑÎ¨¥Í≤ÉÎèÑ Ïïà ÌïòÍ∏∞
        transforms.RandomResizedCrop((224, 224), scale=(0.8, 1.0)),  # (2) ÎûúÎç§ Î¶¨ÏÇ¨Ïù¥Ï¶à ÌÅ¨Î°≠
        transforms.RandomHorizontalFlip(p=1.0),                      # (3) Ï¢åÏö∞ Îí§ÏßëÍ∏∞
        transforms.RandomRotation(degrees=30),                       # (4) ÌöåÏ†Ñ
        transforms.ColorJitter(brightness=0.2, contrast=0.2,
                               saturation=0.2, hue=0.1)              # (5) ÏÉâÏÉÅ Î≥ÄÌôî
    ]),
    # ÌÖêÏÑú Î≥ÄÌôò & Ï†ïÍ∑úÌôî
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# ------------------------------------------------------------------------
# 2) DoubleDataset ÌÅ¥ÎûòÏä§
#    - ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ NÏû• + Ï¶ùÍ∞ï Ïù¥ÎØ∏ÏßÄ NÏû• = Ï¥ù 2NÏû•
# ------------------------------------------------------------------------
class DoubleDataset(datasets.ImageFolder):
    def __init__(self, root, original_transform=None, augmented_transform=None):
        super().__init__(root, transform=None)
        self.original_transform = original_transform
        self.augmented_transform = augmented_transform

    def __len__(self):
        # ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ÏÖã Í∏∏Ïù¥ √ó 2Î∞∞
        return 2 * super().__len__()

    def __getitem__(self, idx):
        # ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ÏÖã Í∏∏Ïù¥
        n = super().__len__()

        if idx < n:
            # [0 ~ n-1] Íµ¨Í∞Ñ ‚Üí ÏõêÎ≥∏ Î≥ÄÌôò
            path, label = self.samples[idx]
            img = self.loader(path)
            if self.original_transform:
                img = self.original_transform(img)
        else:
            # [n ~ 2n-1] Íµ¨Í∞Ñ ‚Üí Ï¶ùÍ∞ï Î≥ÄÌôò
            path, label = self.samples[idx - n]
            img = self.loader(path)
            if self.augmented_transform:
                img = self.augmented_transform(img)

        return img, label

# ------------------------------------------------------------------------
# 3) Îç∞Ïù¥ÌÑ∞ÏÖã ÏÉùÏÑ±
# ------------------------------------------------------------------------
dataset = DoubleDataset(
    root='./train',                # ImageFolder Íµ¨Ï°∞Ïùò ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ Í≤ΩÎ°ú
    original_transform=original_transform,
    augmented_transform=augmented_transform
)

original_len = len(dataset) // 2   # Ïã§Ï†ú ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ Í∞úÏàò
print("ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ Í∞úÏàò:", original_len)
print("2Î∞∞Î°ú ÌôïÏû•Îêú Îç∞Ïù¥ÌÑ∞ÏÖã ÌÅ¨Í∏∞:", len(dataset))

# ------------------------------------------------------------------------
# 4) (ÏÑ†ÌÉù) train/valid/test Î∂ÑÌï†
# ------------------------------------------------------------------------
all_indices = list(range(len(dataset)))
train_idx, tmp_idx = train_test_split(all_indices, test_size=0.2, random_state=42)
val_idx, test_idx = train_test_split(tmp_idx, test_size=0.5, random_state=42)

data_sets = {
    'train': Subset(dataset, train_idx),
    'valid': Subset(dataset, val_idx),
    'test': Subset(dataset, test_idx)
}

# ------------------------------------------------------------------------
# 5) DataLoader Ï†ïÏùò
# ------------------------------------------------------------------------
dataloaders = {
    'train': DataLoader(data_sets['train'],
                        batch_size=params['batch_size'],
                        shuffle=True,
                        num_workers=params['workers'],
                        pin_memory=True),
    'valid': DataLoader(data_sets['valid'],
                        batch_size=params['batch_size'],
                        shuffle=False,
                        num_workers=params['workers'],
                        pin_memory=True),
    'test': DataLoader(data_sets['test'],
                       batch_size=params['batch_size'],
                       shuffle=False,
                       num_workers=params['workers'],
                       pin_memory=True)
}

# ------------------------------------------------------------------------
# 6) Î∞∞Ïπò Ïàò Í≥ÑÏÇ∞ Î∞è Ï∂úÎ†•
# ------------------------------------------------------------------------
batch_num = {
    'train': len(dataloaders['train']),
    'valid': len(dataloaders['valid']),
    'test': len(dataloaders['test'])
}

print('batch_size: %d | batch_num(train/valid/test): %d / %d / %d' %
      (params['batch_size'],
       batch_num['train'],
       batch_num['valid'],
       batch_num['test']))
````

### Train Ìï®Ïàò
**ÏûÖÎ†• Ïù∏Ïûê:**

model: ÌïôÏäµÏãúÌÇ¨ PyTorch Î™®Îç∏.  
criterion: ÏÜêÏã§ Ìï®Ïàò (Ïòà: nn.CrossEntropyLoss).
optimizer: Î™®Îç∏Ïùò ÌååÎùºÎØ∏ÌÑ∞Î•º ÏóÖÎç∞Ïù¥Ìä∏Ìï† ÏòµÌã∞ÎßàÏù¥Ï†Ä (Ïòà: torch.optim.Adam).  
dataloaders: trainÍ≥º valid Îç∞Ïù¥ÌÑ∞Î•º Î°úÎìúÌïòÎäî DataLoader ÎîïÏÖîÎÑàÎ¶¨.  
num_epochs: ÌïôÏäµ ÏóêÌè¨ÌÅ¨ Ïàò (Í∏∞Î≥∏Í∞í: 25).

**Ï∂úÎ†•Í∞í:**

model: ÏµúÏ¢Ö ÌïôÏäµ ÏôÑÎ£åÎêú Î™®Îç∏.  
best_idx: Í∞ÄÏû• ÎÜíÏùÄ Í≤ÄÏ¶ù Ï†ïÌôïÎèÑÎ•º Í∏∞Î°ùÌïú ÏóêÌè¨ÌÅ¨ Î≤àÌò∏.  
best_acc: Í∞ÄÏû• ÎÜíÏùÄ Í≤ÄÏ¶ù Ï†ïÌôïÎèÑ.  
train_loss, train_acc: ÌïôÏäµ ÏÜêÏã§ Î∞è Ï†ïÌôïÎèÑÏùò Í∏∞Î°ù.  
valid_loss, valid_acc: Í≤ÄÏ¶ù ÏÜêÏã§ Î∞è Ï†ïÌôïÎèÑÏùò Í∏∞Î°ù.

````python
def train_model(model, criterion, optimizer, dataloaders: dict, num_epochs=25):
  since = time.time()

  best_model_wts = copy.deepcopy(model.state_dict()) # Í∞ÄÏû• ÎÜíÏùÄ Í≤ÄÏ¶ù Ï†ïÌôïÎèÑÎ•º Í∏∞Î°ùÌïú Î™®Îç∏Ïùò Í∞ÄÏ§ëÏπòÎ•º Ï†ÄÏû•
  best_acc = 0.0
  train_loss, train_acc, valid_loss, valid_acc = [], [], [], []

  for epoch in range(num_epochs): # ÏóêÌè¨ÌÅ¨(epoch) Îã®ÏúÑÎ°ú ÌïôÏäµÏùÑ Î∞òÎ≥µ
    print()
    print('Epoch {}/{}'.format(epoch+1, num_epochs), end=": ")
    print('-'*10)

    # Each epoch has a training and validation phase
    for phase in ['train', 'valid']:
      if phase =='train':
        model.train() # Set model to training mode
      else:
        model.eval() # Set model to evaluate mode

      running_loss, running_corrects, num_cnt = 0.0, 0, 0

      count = 1

      #Iterate over data.
      pbar_dataloaders = tqdm(dataloaders[phase],
                              desc=phase,
                              ncols=70)
      for inputs, labels in pbar_dataloaders:
        inputs = inputs.to(device)
        labels = labels.to(device)

        # zero the parameter gradients
        optimizer.zero_grad() # Ïù¥Ï†Ñ Î∞∞ÏπòÏóêÏÑú Í≥ÑÏÇ∞Îêú Í∑∏ÎùºÎîîÏñ∏Ìä∏Î•º Ï¥àÍ∏∞Ìôî

        # forward
        # track history if only in train
        with torch.set_grad_enabled(phase == 'train'):
          outputs = model(inputs) # Î™®Îç∏Ïóê ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º Ï†ÑÎã¨Ìï¥ Ï∂úÎ†•(logits)ÏùÑ Í≥ÑÏÇ∞
          _, preds = torch.max(outputs, 1) # Ï∂úÎ†• logitsÏóêÏÑú Í∞ÄÏû• ÎÜíÏùÄ Í∞íÏùÑ Í∞ÄÏßÑ ÌÅ¥ÎûòÏä§ ÏòàÏ∏°
          loss = criterion(outputs, labels) # ÏÜêÏã§ Ìï®Ïàò(Ïòà: ÌÅ¨Î°úÏä§ ÏóîÌä∏Î°úÌîº)Î•º ÏÇ¨Ïö©Ìï¥ ÏÜêÏã§ Í≥ÑÏÇ∞

          # backward only if in training phase
          if phase == 'train':
            loss.backward() # Í∑∏ÎùºÎîîÏñ∏Ìä∏Î•º Ïó≠Ï†ÑÌåå(backpropagation)Î°ú Í≥ÑÏÇ∞
            optimizer.step() # Î™®Îç∏Ïùò ÌååÎùºÎØ∏ÌÑ∞Î•º ÏóÖÎç∞Ïù¥Ìä∏

        # statistics
        running_loss += loss.item() * inputs.size(0) # ÏÜêÏã§ Í∞íÏùÑ Î∞∞Ïπò ÌÅ¨Í∏∞ÎßåÌÅº ÎàÑÏ†Å
        running_corrects += torch.sum(preds == labels.data) # Ïò¨Î∞îÎ•¥Í≤å ÏòàÏ∏°Ìïú ÏÉòÌîå ÏàòÎ•º ÎàÑÏ†Å
        num_cnt += len(labels) # Ï≤òÎ¶¨Ìïú ÏÉòÌîå Ïàò(num_cnt) ÎàÑÏ†Å
        count += 1
      pbar_dataloaders.close()

      # ÌòÑÏû¨ ÏóêÌè¨ÌÅ¨ÏóêÏÑúÏùò ÌèâÍ∑† ÏÜêÏã§Í≥º Ï†ïÌôïÎèÑÎ•º Í≥ÑÏÇ∞
      epoch_loss = float(running_loss / num_cnt)
      epoch_acc = float((running_corrects.double() / num_cnt).cpu()*100)

      if phase == 'train':
        train_loss.append(epoch_loss)
        train_acc.append(epoch_acc)
      else:
        valid_loss.append(epoch_loss)
        valid_acc.append(epoch_acc)
      print()
      print(f'{phase} Loss: {epoch_loss:.2f} Acc: {epoch_acc:.1f}')

      # deep copy the model (ÏµúÍ≥† ÏÑ±Îä• Î™®Îç∏ Ï†ÄÏû•)
      if phase == 'valid' and epoch_acc > best_acc:
        best_idx = epoch + 1
        best_acc = epoch_acc
        best_model_wts = copy.deepcopy(model.state_dict())
        print('==> best model saved - %d / %.1f'%(best_idx, best_acc))

  # ÏµúÏ¢Ö Í≤∞Í≥º
  time_elapsed = time.time() - since
  print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
  print('Best valid Acc: idx=%d , %.1f'%(best_idx, best_acc))

  # load best model weights
  model.load_state_dict(best_model_wts)
  torch.save(model.state_dict(), pt_file_name)
  print('model saved')
  return model, best_idx, best_acc, train_loss, train_acc, valid_loss, valid_acc
````

### ÏÜêÏã§Ìï®Ïàò & ÏòµÌã∞ÎßàÏù¥Ï†Ä

Adam:

## Î™®Îç∏ TEST

````python
def test_and_visualize_model(model, dataloaders, phase = 'test', num_images=4):
  # phase = 'train', 'valid', 'test'

  was_training = model.training
  model.eval()

  running_loss, running_corrects, num_cnt = 0.0, 0, 0

  with torch.no_grad():
    for i, (inputs, labels) in enumerate(dataloaders[phase]):
      inputs = inputs.to(device)
      labels = labels.to(device)

      outputs = model(inputs)
      _, preds = torch.max(outputs, 1)
      loss = criterion(outputs, labels) # batchÏùò ÌèâÍ∑† loss Ï∂úÎ†•

      running_loss += loss.item() * inputs.size(0)
      running_corrects += torch.sum(preds == labels.data)
      num_cnt += inputs.size(0) # batch size

    test_loss = running_loss / num_cnt
    test_acc = running_corrects.double() / num_cnt
    print('test done : loss/acc : %.2f / %.1f' % (test_loss, test_acc*100))

  # ÏòàÏãú Í∑∏Î¶º Ï∂úÎ†•
  with torch.no_grad():
    for i, (inputs, labels) in enumerate(dataloaders[phase]):
      inputs = inputs.to(device)
      labels = labels.to(device)

      outputs = model(inputs)
      _, preds = torch.max(outputs, 1)

      for j in range(1, num_images+1):
        ax = plt.subplot(num_images//2, 2, j)
        ax.axis('off')
        ax.set_title('%s : %s -> %s' %(
          'True' if class_names[str(labels[j].cpu().numpy())] == class_names[str(preds[j].cpu().numpy())] else 'False',
          class_names[str(labels[j].cpu().numpy())],
          class_names[str(preds[j].cpu().numpy())]
        ))
        imgShow(inputs.cpu().data[j])

      if i == 0: break

  model.train(mode=was_training) # Îã§Ïãú trainÎ™®ÎìúÎ°ú Ï†ÑÌôò
````

````python
test_and_visualize_model(model, dataloaders, phase='test')
````
